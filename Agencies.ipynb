```python
test_A['all_purchases'] = test_A["post_click_conv"] + test_A['post_view_conv']
test_B['all_purchases'] = test_B["post_click_conv"] + test_B['post_view_conv']
```

```python
all_uniq_A = test_A['campaign_group_id'].unique()
all_uniq_B = test_B['campaign_group_id'].unique()
similar = []
for a in all_uniq_A:
    if a in all_uniq_B:
        similar.append(a)
#13 similar groups
```

```python
sim_test_A = test_A[test_A['campaign_group_id'].isin(similar)] #84 rows
sim_test_B = test_B[test_B['campaign_group_id'].isin(similar)] #86 rows
sim_test_A_camp = sim_test_A.groupby('campaign_group_id')
sim_test_B_camp = sim_test_B.groupby('campaign_group_id')
```

```python
c_for_g_A = sim_test_A_camp['clicks'].sum()
c_for_g_B = sim_test_B_camp['clicks'].sum()
c_for_g_A.sort_index()
c_for_g_B.sort_index()
```

```python
c_for_g_A = c_for_g_A.reset_index()
c_for_g_B = c_for_g_B.reset_index()
```

```python
c_for_g_A['all_clicks'] = ((c_for_g_B['clicks'] - c_for_g_A['clicks'])*100)/c_for_g_A['clicks']
```

```python
sns.set(style="whitegrid")
ax = sns.barplot(x=c_for_g_A.index, y='all_clicks', data = c_for_g_A)#, palette=['C3', 'C9'])
```
